```python
def plot_corr (df,size=11):
    corr=df.corr()
    fig, ax = plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    cax = ax.matshow(corr, interpolation='nearest')
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)),corr.columns,rotation=17) #draw xticks
    plt.yticks(range(len(corr.columns)),corr.columns) #draw yticks
```

```python
plot_corr(df)
```

```python
#diabet değişkenine 1 ve 0 değerlerini atadık
diabetes_map={True:1,False:0}
df['diabetes']=df['diabetes'].map(diabetes_map)
```

```python
df.head(5)
```

```python
#elimizde bulunan true ve false'ların yüzdelik değerleri
num_true=len(df.loc[df['diabetes']==True])
num_false=len(df.loc[df['diabetes']==False])

print("Total num of true = {0} " .format(num_true,num_true/(num_true+num_false)*100))
print("Total num of true = {0} " .format(num_false,num_false/(num_false+num_true)*100))
```

```python
df.columns
```

```python
from sklearn.cross_validation import train_test_split

feature_col_names = [ 'glucose_conc', 'diastolic_bp', 'thickness', 'insulin',
       'bmi', 'diab_pred', 'age']
predicted_class_names =['diabetes']

X=df[feature_col_names].values
y=df[predicted_class_names].values

split_test_size=0.30

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=split_test_size,random_state=42)
```

```python
print("Training Data : {0:2.2f}%".format(len(X_train)/len(df)*100))
print("Test Data : {0:2.2f}%".format(len(X_test)/len(df)*100))
```

```python
#bütün değişkenlerde ki O değerlenini sayısını gösteriyor
for column in df.columns:
    print("Number of zeroes in {1:12} :  {0:>4}".format(len(df.loc[df[column]==0]),column))
```

```python
from sklearn.preprocessing import Imputer
```

```python
X_train[45] #45 satırda ki 0 değerini görüyorsunuz
```

```python
#Imputer O değerlerini ortalama  bir değer ekliyoruz
fill_O=Imputer(missing_values=0,strategy='mean',axis=0)
#axis=0 colon,axis=1 satır

X_train=fill_O.fit_transform(X_train)
X_test=fill_O.fit_transform(X_test)
```

```python
X_train[25]
```

```python
y_train.ravel()
```

```python
#Naibe bayes modelini import ediyoruz
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
nb_model.fit(X_train,y_train.ravel())
```

```python
#predict eğitim verisi
nb_predict_train=nb_model.predict(X_train)

#metric kütüphanesini import ediyoruz

from sklearn import metrics

print("Accuracy : {0:2.4f}".format(metrics.accuracy_score(y_train,nb_predict_train)) )
```

```python
#y_test scoreb
nb_predict_test=nb_model.predict(X_test)
print("Accuracy : {0:2.4f}".format(metrics.accuracy_score(y_test,nb_predict_test)))
```

```python
conf_mx=metrics.confusion_matrix(y_test,nb_predict_test,labels=[1,0])
```
