```python
def plot_corr (hrm,size=11):
    corr=hrm.corr()
    fig,ax=plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    cax=ax.matshow(corr,interpolation='nearest')
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)),corr.columns,rotation=100)
    plt.yticks(range(len(corr.columns)),corr.columns)
```

```python
plot_corr(hrm)
```

```python
from sklearn.cross_validation import train_test_split

feature_col_names=[u'Age', u'BusinessTravel', u'DailyRate', u'Department',
       u'DistanceFromHome', u'Education', u'EducationField', u'EmployeeCount',
       u'EmployeeNumber', u'EnvironmentSatisfaction', u'Gender', u'HourlyRate',
       u'JobInvolvement', u'JobLevel', u'JobSatisfaction', u'MaritalStatus',
       u'MonthlyIncome', u'MonthlyRate', u'NumCompaniesWorked', u'OverTime',
       u'PercentSalaryHike', u'PerformanceRating', u'RelationshipSatisfaction',
       u'StandardHours', u'StockOptionLevel', u'TotalWorkingYears',
       u'TrainingTimesLastYear', u'WorkLifeBalance', u'YearsAtCompany',
       u'YearsInCurrentRole', u'YearsSinceLastPromotion',
       u'YearsWithCurrManager']

predicted_class_names=['Attrition']

X=hrm[feature_col_names].values
y=hrm[predicted_class_names].values

split_test_size=0.30

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=split_test_size,random_state=42)
```
